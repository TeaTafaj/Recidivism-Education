# Data Exploration – National Corrections Reporting Program (NCRP)

### Dataset Source
National Corrections Reporting Program, 1991–2020: Selected Variables  
(ICPSR 38492)

This dataset provides individual-level prison records across U.S. states over multiple decades, including information on inmate demographics, offense types, admission and release timing, and incarceration characteristics.

---

## Dataset Structure

The dataset is organized into multiple subfolders, each containing a tab-separated data file (`.tsv`) representing a different component of the NCRP data.

- **DS0001** – Inmate-level records including admission, release, and demographic characteristics
- **DS0002** – Records organized by reporting year with inmate and offense characteristics
- **DS0003** – Prison release records with information on release year, release type, and time served
- **DS0004** – prison population at the end of each year*

Each dataset contains structured tab-separated files that can be loaded and explored using Python.

---

## Exploration Goal

Below, we quickly inspect each dataset to understand:

- what information each file contains  
- which variables may be useful for our project  
- which data we still need but may be missing  
- how these datasets could potentially be combined or aggregated for analysis

**DS-01**

```python
import pandas as pd

df = pd.read_csv("DS0001/38492-0001-Data.tsv", sep="\t")
df.head()
```

**Dataset Exploration – DS0001 (Inmate Characteristics)**

Each row in this dataset represents an individual inmate observation.

### Variables

- **ABT_INMATE_ID** – Unique identifier for each inmate
- **SEX** – Sex of the inmate
- **ADMTYPE** – Admission type to prison
- **OFFGENERAL** – General category of offense
- **OFFDETAIL** – Detailed offense category
- **EDUCATION** – Education level of the inmate
- **ADMITYR** – Year the inmate was admitted to prison
- **RELEASEYR** – Year the inmate was released from prison
- **MAND_PRISREL_YEAR** – Mandatory prison release year
- **PROJ_PRISREL_YEAR** – Projected prison release year
- **PARELIG_YEAR** – Year the inmate becomes eligible for parole
- **SENTLGTH** – Length of the sentence
- **RACE** – Race of the inmate
- **AGEADMIT** – Age at admission
- **AGERELEASE** – Age at release
- **TIMESRVD** – Time served in prison
- **RELTYPE** – Type of release from prison
- **STATE** – State where the incarceration occurred


### Purpose of this Dataset

This dataset contains individual-level information on incarcerated individuals, including demographics, offense characteristics, admission and release timing, and sentence details.

Key variables for our analysis include:
- **STATE** – allows aggregation to the state level
- **ADMITYR / RELEASEYR** – allow construction of timelines
- **SENTLGTH and TIMESRVD** – indicate incarceration duration
- **RELTYPE** – may help identify release conditions

This dataset may help us construct measures related to incarceration patterns and prisoner characteristics across states and years, which could be useful when linking prison populations to recidivism outcomes.
