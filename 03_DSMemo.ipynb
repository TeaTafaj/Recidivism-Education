# Stakeholder

This analysis is intended for **state legislators (House and Senate) and Governors** responsible for designing, approving, and funding correctional policies.

These stakeholders determine:
- The structure and funding of prison education programs  
- The adoption of policies such as sentence reduction incentives, automatic enrollment, and centralized administration  
- The broader strategy for reducing recidivism and improving post-release outcomes  

The goal of this memo is to inform policy decisions by identifying which education-related policies are most effective in reducing recidivism.

________

# Executive Summary

The United States incarcerates approximately 1.8–1.9 million individuals, with over 1.25 million held in state and federal prisons alone, representing the highest incarceration rate globally. High recidivism rates place sustained economic and social strain on communities, families, and correctional systems.

A central objective of incarceration is rehabilitation, with recidivism—defined here as a return to prison after release—serving as a key measure of success. While education programs have historically been used to support rehabilitation, their effectiveness varies depending on how they are implemented across states.

This analysis evaluates how education-related policies influence recidivism, focusing on two approaches: the expansion of educational access through the Pell Second Chance Program (post-2016), and variation in state-level policy design, including central administration, automatic enrollment, and sentence reduction incentives.

- **Pell Second Chance Program:** A policy that restores access to federal Pell Grants, allowing incarcerated individuals to enroll in postsecondary education programs while in prison.

- **Sentence Reduction Incentives:** Policies that reduce an individual’s prison sentence in exchange for successful completion of educational programs.

- **Central Administration:** A policy structure where prison education programs are managed through a centralized, statewide system or governing body.

- **Automatic Enrollment:** Policies that automatically enroll eligible incarcerated individuals into educational programs based on literacy level or educational attainment.

The results indicate that expanded access to education through Pell is associated with an estimated **3.3 percentage point reduction in recidivism**. However, policy design plays a more substantial role: among the policies examined, **sentence reduction incentives are associated with a 6.6 percentage point reduction in recidivism**, representing the largest observed effect.

These findings suggest that while education contributes to rehabilitation, **policies that directly incentivize participation and completion of educational programs are significantly more effective in reducing recidivism**.

# Policy Recommendations

Based on the findings, policy design plays a critical role in determining the effectiveness of prison education programs in reducing recidivism. While expanding access to education is associated with improved outcomes, policies that actively incentivize participation yield substantially stronger effects.

### 1. Prioritize Sentence Reduction Incentives
- Implement or expand policies that reduce prison sentences in exchange for completion of educational programs  
- These policies are associated with an estimated **6.6 percentage point reduction in recidivism**, the largest observed effect among the policies analyzed  
- Incentive structures appear to directly motivate engagement and completion, leading to stronger rehabilitation outcomes  

### 2. Maintain and Expand Access to Educational Programs
- Continue supporting programs such as the **Pell Second Chance Program**, which are associated with reductions in recidivism  
- Expanding access ensures that incarcerated individuals have the opportunity to participate in education programs  

### 3. Combine Access with Incentive-Based Design
- Education access alone is not sufficient to maximize impact  
- Policies should be structured to **pair access with clear incentives**, reinforcing participation and completion  

### 4. Deprioritize Standalone Structural Policies
- Policies such as central administration or automatic enrollment, when implemented without incentives, show weaker or negligible associations with recidivism reduction  
- These policies may support system organization but should not be relied upon as primary tools for reducing recidivism
