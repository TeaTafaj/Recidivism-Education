# Stakeholder

This analysis is intended for **state legislators (House and Senate) and Governors** responsible for designing, approving, and funding correctional policies.

These stakeholders determine:
- The structure and funding of prison education programs  
- The adoption of policies such as sentence reduction incentives, automatic enrollment, and centralized administration  
- The broader strategy for reducing recidivism and improving post-release outcomes  

The goal of this memo is to inform policy decisions by identifying which education-related policies are most effective in reducing recidivism.

________

# Executive Summary

The United States incarcerates approximately 1.8–1.9 million individuals, with over 1.25 million held in state and federal prisons alone, representing the highest incarceration rate globally. High recidivism rates place sustained economic and social strain on communities, families, and correctional systems.

A central objective of incarceration is rehabilitation, with recidivism—defined here as a return to prison after release—serving as a key measure of success. While education programs have historically been used to support rehabilitation, their effectiveness varies depending on how they are implemented across states.

This analysis evaluates how education-related policies influence recidivism, focusing on two approaches: the expansion of educational access through the Pell Second Chance Program (post-2016), and variation in state-level policy design, including central administration, automatic enrollment, and sentence reduction incentives.

- **Pell Second Chance Program:** A policy that restores access to federal Pell Grants, allowing incarcerated individuals to enroll in postsecondary education programs while in prison.

- **Sentence Reduction Incentives:** Policies that reduce an individual’s prison sentence in exchange for successful completion of educational programs.

- **Central Administration:** A policy structure where prison education programs are managed through a centralized, statewide system or governing body.

- **Automatic Enrollment:** Policies that automatically enroll eligible incarcerated individuals into educational programs based on literacy level or educational attainment.

The results indicate that expanded access to education through Pell is associated with an estimated **3.3 percentage point reduction in recidivism**. However, policy design plays a more substantial role: among the policies examined, **sentence reduction incentives are associated with a 6.6 percentage point reduction in recidivism**, representing the largest observed effect.

These findings suggest that while education contributes to rehabilitation, **policies that directly incentivize participation and completion of educational programs are significantly more effective in reducing recidivism**.

# Policy Recommendations

Based on the findings, policy design plays a critical role in determining the effectiveness of prison education programs in reducing recidivism. While expanding access to education is associated with improved outcomes, policies that actively incentivize participation yield substantially stronger effects.

### 1. Prioritize Sentence Reduction Incentives
- Implement or expand policies that reduce prison sentences in exchange for completion of educational programs  
- These policies are associated with an estimated **6.6 percentage point reduction in recidivism**, the largest observed effect among the policies analyzed  
- Incentive structures appear to directly motivate engagement and completion, leading to stronger rehabilitation outcomes  

### 2. Maintain and Expand Access to Educational Programs
- Continue supporting programs such as the **Pell Second Chance Program**, which are associated with reductions in recidivism  
- Expanding access ensures that incarcerated individuals have the opportunity to participate in education programs  

### 3. Combine Access with Incentive-Based Design
- Education access alone is not sufficient to maximize impact  
- Policies should be structured to **pair access with clear incentives**, reinforcing participation and completion  

### 4. Deprioritize Standalone Structural Policies
- Policies such as central administration or automatic enrollment, when implemented without incentives, show weaker or negligible associations with recidivism reduction  
- These policies may support system organization but should not be relied upon as primary tools for reducing recidivism  



# Context and Background

Prison education programs have existed for centuries, evolving in structure and purpose over time. Early efforts in the 18th and 19th centuries focused on religious instruction and basic literacy, aimed at moral reform and reintegration. By the early 20th century, formal educational programs began to emerge within prisons, including the first college-level courses. A significant shift occurred in the 1930s, when vocational training became central to rehabilitation efforts, reflecting the belief that equipping individuals with practical skills would improve employment outcomes and reduce recidivism. Today, prison education encompasses a wide range of programs, including vocational training, postsecondary education, and life skills development, with substantial variation in implementation across states.

The role of rehabilitation within the prison system has long been debated. While education programs were initially developed as tools for reform, this perspective has been challenged by periods emphasizing punishment over rehabilitation. This tension is reflected in both policy and research. Access to postsecondary education in prisons was significantly reduced in 1994, when Pell Grant eligibility for incarcerated individuals was removed, leading to a decline in education programs nationwide. In contrast, more recent policy efforts, such as the introduction of the Second Chance Pell initiative in 2015–2016, reflect a renewed emphasis on rehabilitation by expanding access to education within prisons.

This divide is also evident in the academic literature. Early work, most notably Martinson (1974), questioned whether rehabilitation programs meaningfully reduced recidivism, contributing to skepticism about their effectiveness. Later research, including the Prison Education Research Project (1994), found that participation in educational and vocational programs is associated with lower recidivism, improved employment outcomes, and better institutional behavior.

While existing research largely supports the role of education in improving rehabilitation outcomes, it primarily focuses on whether education works, rather than how it is implemented. Given the variation in policy design across states, there remains an open question as to which specific policy structures are most effective in reducing recidivism. This analysis addresses that gap by examining how different education-related policies influence reoffending outcomes.

# Data and Approach

This analysis uses individual-level data from the National Corrections Reporting Program (NCRP), which tracks prison admissions and releases across U.S. states. Recidivism is defined as a return to prison within two years of release. To understand how education policy affects recidivism, the analysis follows two approaches: first, it examines the implementation of the Pell Second Chance Program as an expansion of access to college-level education in prison; second, it evaluates how variation in state-level policy design is associated with recidivism outcomes.

## Methodology 1: Pell Second Chance Program

To evaluate the impact of expanded access to education, this analysis examines the implementation of the Pell Second Chance Program, which restored access to college-level education for incarcerated individuals beginning in 2016.

The analysis compares recidivism outcomes before and after this policy change between states that adopted Pell-funded education programs and those that did not. This allows us to understand how increased access to postsecondary education is associated with changes in recidivism over time.

![Pell Recidivism Plot](pell_plot.png)

In raw terms, recidivism declines by approximately **1.4 percentage points** in states without Pell access, compared to a **11.8 percentage point decrease** in states with Pell-funded programs. This highlights the substantially larger reduction observed in states with expanded access to postsecondary education.

To ensure that these comparisons are not driven by differences in the populations being released across states and over time, the analysis accounts for key individual characteristics. Specifically, we control for factors such as age at admission, gender, sentence length, and type of offense.

This allows the comparison to focus on individuals with similar profiles, reducing the likelihood that differences in recidivism are driven by underlying demographic or criminal characteristics rather than access to education.

## Methodology 2: State Policy Design

To understand how different approaches to prison education influence recidivism, this analysis examines variation in policy design across states. Rather than focusing only on access to education, this approach evaluates how specific policy structures are associated with recidivism outcomes.

Across the policy types examined, we observe reductions in recidivism associated with education-focused policies. In particular, sentence reduction incentives are associated with an estimated **6–7 percentage point decrease** in recidivism, while automatic enrollment and centralized administration show smaller effects.

Given this variation, we focus on **sentence reduction incentives** as the primary policy of interest.

![Sentence Reduction Effect](sentence_reduction_plot.png)

States that offer sentence reduction incentives exhibit approximately **17% lower recidivism** compared to states without such incentives. 

This corresponds to a reduction of about **5 percentage points**, indicating that incentive-based education policies are associated with meaningfully improved post-release outcomes.

Additional Explorations:

Consistent with prior evidence, recidivism decreases with age at admission, indicating that younger individuals face higher risks of reoffending. 

We also observe that female individuals have lower recidivism rates compared to males. 

Recidivism rates vary across offense types, reinforcing the importance of accounting for differences in criminal behavior when evaluating policy impacts.
