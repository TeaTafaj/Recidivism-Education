```python
import gdown
import pandas as pd
import os

file_id = "1jkBkjOsdr-0hhUgj3p2J7pk1Welkk7A1"
url = f"https://drive.google.com/uc?id={file_id}"
output_dir = "../data/raw"
os.makedirs(output_dir, exist_ok=True)
output = os.path.join(output_dir, "38492-0001-Data.tsv")

# Remove the file if it exists to force re-download
if os.path.exists(output):
    os.remove(output)

gdown.download(url, output, quiet=False)

df = pd.read_csv(output, sep="\t", low_memory=False)
print(df.head())
```
