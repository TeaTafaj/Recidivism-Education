```python
import gdown
import pandas as pd
import os
```

```python
# Download
file_id = "1jkBkjOsdr-0hhUgj3p2J7pk1Welkk7A1"
raw_dir = "../data/raw"
os.makedirs(raw_dir, exist_ok=True)
tsv_path = f"{raw_dir}/NCRP_1991_2020.tsv"

if not os.path.exists(tsv_path):  # skip re-download if already present
    gdown.download(f"https://drive.google.com/uc?id={file_id}", tsv_path, quiet=False)

# Read & inspect
df = pd.read_csv(tsv_path, sep="\t", low_memory=False)

print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
df.head()
```

```python
# Read & inspect
parquet_path = "../data/processed/NCRP_1991_2020.parquet"
os.makedirs("../data/processed", exist_ok=True)
df.to_parquet(parquet_path, index=False)
print(f"Saved to {parquet_path}")
```
