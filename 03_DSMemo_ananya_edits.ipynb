# Executive Summary

The United States incarcerates approximately 1.8–1.9 million individuals, with over 1.25 million held in state and federal prisons alone, representing the highest incarceration rate globally. High recidivism rates place sustained economic and social strain on communities, families, and correctional systems. A central objective of incarceration is rehabilitation, with less recidivism—defined here as a return to prison after release—serving as a key measure of success. While education programs have historically been used to support rehabilitation, their effectiveness varies depending on how they are implemented across states.

This analysis evaluates how education-related policies influence recidivism, focusing on two approaches: the expansion of educational access through the Pell Second Chance Program (post-2016), and variation in state-level policy design, including central administration, automatic enrollment, and sentence reduction incentives.

- **Pell Second Chance Program:** A policy that restores access to federal Pell Grants, allowing incarcerated individuals to enroll in postsecondary education programs while in prison.

- **Sentence Reduction Incentives:** Policies that reduce an individual’s prison sentence in exchange for successful completion of educational programs.

- **Central Administration:** A policy structure where prison education programs are managed through a centralized, statewide system or governing body.

- **Automatic Enrollment:** Policies that automatically enroll eligible incarcerated individuals into educational programs based on literacy level or educational attainment.

The results indicate that policy design plays the most substantial/effective role in reducing recidivism: **sentence reduction incentives are associated with a 6.6 percentage point reduction in recidivism**, representing the largest observed effect. Apart from this, expanded access to education through the Pell program is also associated with an estimated **3.3 percentage point reduction in recidivism**. 

These findings suggest that while education contributes to rehabilitation, **policies that directly incentivize participation in and completion of educational programs are significantly more effective in reducing recidivism**.

# Context and Background

Prison education programs have existed for centuries, evolving in structure and purpose over time. Early efforts in the 18th and 19th centuries focused on religious instruction and basic literacy, aimed at moral reform and reintegration. By the early 20th century, formal educational programs began to emerge within prisons, including the first college-level courses. A significant shift occurred in the 1930s, when vocational training became central to rehabilitation efforts, reflecting the belief that equipping individuals with practical skills would improve employment outcomes and reduce recidivism. Today, prison education encompasses a wide range of programs, including vocational training, postsecondary education, and life skills development, with substantial variation in implementation across states.

The role of rehabilitation within the prison system has long been debated. While education programs were initially developed as tools for reform, this perspective has been challenged by periods emphasizing punishment over rehabilitation. This tension is reflected in the research literature: early work, most notably Martinson (1974), questioned whether rehabilitation programs meaningfully reduced recidivism, contributing to skepticism about their effectiveness. In contrast, later studies, including the Prison Education Research Project (1994), found that participation in educational and vocational programs is associated with lower recidivism, improved employment outcomes, and better institutional behavior.

More recently, policy efforts such as the expansion of the Pell Second Chance Program have renewed focus on education as a rehabilitation tool by increasing access to postsecondary programs within prisons. However, while access to education has expanded, the structure and implementation of education policies continue to vary significantly across states.

While existing research largely supports the role of education in improving rehabilitation outcomes, it primarily focuses on whether education works, rather than how it is implemented. Given the variation in policy design across states, there remains an open question as to which specific policy structures are most effective in reducing recidivism. This analysis addresses that gap by examining how different education-related policies influence reoffending outcomes.
